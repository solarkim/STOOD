```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
from utils.plots import plot_bar
from config import PATH_LOG, PATH_MODEL

CWD = Path.cwd()

OOD_LOG_DIR = PATH_LOG / 'saved_models'
```

## IN - OOD [Cifar10 - Texture]

```python
eval_results = {'ms_resnet50': 'cifar10_ms_resnet50_23-06-22T09:42:24/evaluation_ood_models_cifar10_ms_resnet50.csv',
                'resnet50': 'cifar10_resnet50_23-06-22T09:42:24/evaluation_ood_models_cifar10_resnet50.csv',
                'vit16': 'cifar10_vit16_23-06-22T11:01:34/evaluation_ood_models_cifar10_vit16.csv',
}
```

```python
dfs = pd.DataFrame()

for file in eval_results.values():
    df = pd.read_csv(OOD_LOG_DIR / file)
    dfs = pd.concat([dfs, df])

for model in dfs.Model.unique():
    df2 = dfs.loc[dfs.Model == model, 'Detector': 'ACC'].drop('Model', axis=1)
    plot_bar(df2, col_x='Detector', layout=(1,5), title=model, x_rot=45, figsize=(20,4), sharey=True)
    plt.subplots_adjust(wspace=0.1)  # Adjust the spacing between subplots
```
